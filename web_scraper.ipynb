```python
import requests
import pandas
from bs4 import BeautifulSoup

r = requests.get("https://wealthygorilla.com/top-20-richest-people-world/")
c = r.content

soup = BeautifulSoup(c, "html.parser")


l = []
all = soup.find_all("strong")
for items in all:
    d = {}
    new = items.find_all("a")
    if len(new) == 0:
        ans = items.text
        if ans.find("Net") == 0:
            global sol
            sol = ans.replace("Net Worth: ", "")
            
    if len(new) == 1:
        d["Name"] = items.find("a").text
        d["Net Worth"] = sol

    l.append(d)
    
    
for items in l:
    if len(items) > 0:
        print(items)
          
z = []
for items in l:
    if len(items) > 0:
        print(items)
        z.append(items)

 
mf = pandas.DataFrame(z)
```
